```python
performance_df["Gender"] = performance_df["Gender"].map({0:'Male', 1: 'Female'})
```

```python
performance_df["Ethnicity"]= performance_df["Ethnicity"].map({0:'Caucasian', 1:'African Amercian', 2: 'Asian', 3:'Other'})
```

```python
performance_df["ParentalEducation"] = performance_df["ParentalEducation"].map({0:'None', 1:'High School', 2:'Some College', 3:'Bachelor\'s', 4:'Higher'})
```

```python
performance_df["Tutoring"] = performance_df["Tutoring"].map({0:'No', 1:'Yes'})
```

```python
performance_df["Extracurricular"] = performance_df["Extracurricular"].map({0:'No', 1:'Yes'})
```

```python
performance_df["Sports"] = performance_df["Sports"].map({0:'No', 1:'Yes'})
```

```python
performance_df["Music"] = performance_df["Music"].map({0:'No', 1:'Yes'})
```

```python
performance_df["Volunteering"] = performance_df["Volunteering"].map({0:'No', 1:'Yes'})
```

```python
performance_df["GradeClass"] = performance_df["GradeClass"].map({0.0:'A', 1.0:'B', 2.0:'C', 3.0:'D', 4.0:'F'})
```

```python
performance_df.head(10)
```

```python
performance_df = performance_df.round({'StudyTimeWeekly': 1, 'GPA':2})
```

```python
i = np.arange(5)
plt.pyplot.figure(figsize=(20,8))
width = 0.10
multiplier = 0

plt.pyplot.bar(i, parental_edu_none_list,width, label = 'None')
plt.pyplot.bar(i + width, parental_edu_highschool_list,width, label = 'High School')
plt.pyplot.bar(i + (2*width), parental_edu_somecollege_list,width, label = 'Some College')
plt.pyplot.bar(i + (3*width), parental_edu_bachelors_list,width, label = 'Bachelors')
plt.pyplot.bar(i + (4*width), parental_edu_higher_list,width, label = 'Higher')

plt.pyplot.legend(loc = 'best')
plt.pyplot.xticks(i + width, list_gradeclass)
plt.pyplot.show()
```
